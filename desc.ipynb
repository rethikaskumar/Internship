```python
for s in fileList:
    print(s)
    dct={}
    df=pd.read_csv(s, low_memory=False)
    colList=df.columns
    #print(colList)
    for i in colList:
        if re.search(r'(?i)ID$', i) or re.search(r'(?i)zip', i) or re.search(r'(?i)month', i) or re.search(r'(?i)code', i):         #covid
            if i !='Covid':
                #print(i)
                df[i]=df[i].astype(str)
                print(df[i].dtype)
        if re.search(r'(?i)year', i) or re.search(r'(?i)date', i) or i=='published_at':
            if i!='Two Year Change' and s!='loan_data.csv' and s!='World GDP.csv':
                #print(i)
                df[i]=pd.to_datetime(df[i], format='mixed')
                print(df[i].dtype)
    file=s
    for i in df.columns:
        if i=='volume':
            print(df[i].dtype)
        #print(type(df.loc[0,i]))
        dct[i]=df[i].describe()
        #print(type(dct[i][0]))
        if dct[i].dtype==np.float64:
            dct[i]['kurtosis']=dct[i].kurtosis()
            dct[i]=dct[i].round(2)
        if df[i].dtype==np.int_ or df[i].dtype==np.float_ or df[i].dtype==np.int64:
            print(i)
            dct[i]['type']='measure'
            n=len(df[i])
            #print(df[i])
            #print(df[i].value_counts())
            if df[i].dtype==np.int_:
                if 0 in df[i].value_counts():
                    z=df[i].value_counts()[0]
                else:
                    z=0
            elif df[i].dtype==np.float_:
                df[i]=df[i].round(2)
                if 0.00 in df[i].value_counts():
                    z=(df[i].value_counts()[0.00])
                else:
                    z=0
            else:
                z=0
            dct[i]['No. Zero']=str(z)+'('+str(round((z/n)*100, 2))+'%)'
            nz=n-z
            dct[i]['No. Non Zero']=str(nz)+'('+str(round((nz/n)*100, 2))+'%)'
            hist, bins=np.histogram(df[i].dropna().values, bins=3)
            bins=bins.round(2)
            histdf={}
            print(hist)
            for k in range(len(bins)-1):
                histdf[str(bins[k])+' - '+str(bins[k+1])]=str(hist[k])+'('+str(round((hist[k]/sum(hist))*100, 2))+'%)'
            dct[i]['charData']=histdf
            
        else:
            if re.search(r'(?i)ID$', i) or re.search(r'(?i)zip', i) or re.search(r'(?i)month', i) or re.search(r'(?i)code', i):         #covid
                if i !='Covid':
                    if len(set(df[i]))==len(df[i]):
                        dct[i]['type']='Unique ID'
                    else:
                        dct[i]['type']='ID'
            elif re.search(r'(?i)year', i) or re.search(r'(?i)date', i) or i=='published_at':
                if i!='Two Year Change' and s!='loan_data.csv' and s!='World GDP.csv':
                    dct[i]['type']='Date'
                    dct[i]['range']=dct[i]['max']-dct[i]['min']
            else:
                dct[i]['type']='dimension'
        dct[i]=str(dct[i].to_dict())
        #print(dct[i])
    dfdict[file]=dct
    #dfdict[s]=df.select_dtypes(include='number').describe().to_dict()
```
