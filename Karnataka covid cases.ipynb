```python
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0 (X11; Windows; Windows x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.114 Safari/537.36')]
urllib.request.install_opener(opener)
urllib.request.urlretrieve(url, "01-07-2020.pdf")
```

```python
from PyPDF2 import PdfReader
```

```python
import re
```

```python
reader=PdfReader('01-07-2020.pdf')
```

```python
page=reader.pages[1]
```

```python
text=page.extract_text()
```

```python
list1=re.split('\n[0-9]{1,2}', text)[1:]
```

```python
list1[-1]=list1[-1].split('\n')[0]
```

```python
list2=[]
```

```python
list1
```

```python
for i in list1:
    i=re.sub('([a-z])\s([A-z])', '\g<1>\g<2>', i).strip(' ')
    i=i.replace("  ", " ")
    list2.append(i.split(" "))
```

```python
list2
```

```python
tdf=pd.DataFrame(list2)
```

```python
tdf.iloc[:, 1:]=tdf.iloc[:, 1:].astype(int)
```

```python
tdf.columns=cols
```
